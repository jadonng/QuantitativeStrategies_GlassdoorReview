```python
df = pd.read_parquet("text-reviews.parquet")
df
```

```python
# get tickers
tickers = dict(df["ticker"])
reviewId = dict(df["reviewId"])
reviewDateTime = dict(df["reviewDateTime"])
# print(type(tickers))
# print(tickers)
# print(reviewId)
```

```python
df0 = pd.read_parquet("fromLocal/final.parquet", engine='fastparquet')
df0
```

```python
df0[df0['index']==1000]
```

average out all the same metric sentiment value

```python
# Group by 'id', 'name', and 'category' and calculate the mean of 'value'
df_grouped = df0.groupby(['Review', 'Metric', 'index'], as_index=False)['Sentiment Score'].mean()
# df_grouped.sort_values(by='index', inplace=True)
df_grouped[df_grouped['index'] == 1000]
```

compile the average df into new columns on every line of comments

```python
# Create a pivot table
df_pivot = df_grouped.pivot_table(index=['index', 'Review'], columns='Metric', values='Sentiment Score', aggfunc='sum', fill_value=0)
df_pivot.head(1)
```

merge the df with the original and take the thing 

completed for loop process

```python
df_out = df_pivot
```

```python
df_out.reset_index(inplace=True)
df_out
```

```python
df_out['ticker'] = df_out['index'].map(tickers)
df_out['reviewId'] = df_out['index'].map(reviewId)
df_out['reviewDateTime'] = df_out['index'].map(reviewDateTime)
df_out
```

```python
df_out.to_parquet('final/backtest-ready-full.parquet', engine='fastparquet')
```
